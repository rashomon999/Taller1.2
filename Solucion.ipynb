```python
# @title $7) $ Let $a_n = \frac{10^n}{n!} $ for n=1,2,3,....100.
import matplotlib.pyplot as plt
import numpy as np
import math

# Definir los valores de n
n_values = np.arange(1, 100)  # Generar valores de 1 a 199

# Calcular la secuencia a_n
a_n = [(10 ** n) / math.factorial(n) for n in n_values]

# Graficar la secuencia a_n
#plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n,'.', label='a_n = 10^n / n!', color='blue')
plt.title('Secuencia a_n = 10^n / n!')
plt.xlabel('n')
plt.ylabel('Valor de la secuencia')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# @title $8)$ Consider the sequence {𝑎𝑛} with $a_1 = 10 $ and $ a_{n+1}=\frac{1}{3} a_n$  $$$$ for 𝑛 = 1, 2, 3, .. Plot the points (𝑛, 𝑎𝑛) for 𝑛 = 1, 2, 3, ... , 50. Do you think this sequence has a limit? Can you verify this?

import matplotlib.pyplot as plt

# Definir el número de términos a calcular
num_terms = 50

# Inicializar la secuencia
a_n = [10]  # a_1 = 10

# Calcular los términos siguientes
for n in range(1, num_terms):
    a_n.append(a_n[-1] / 3)

# Definir los valores de n
n_values = list(range(1, num_terms + 1))

# Graficar la secuencia
plt.grid(True)
plt.plot(n_values, a_n, '.', color='blue', label='$a_n = 10 \cdot (1/3)^{n-1}$')
plt.legend()
plt.show()
```

$a_{n+1} = \frac{1}{3} a_n $

inicialmente $a_1$ = 10

$a_2 = \frac{1}{3} * 10 = \frac{10}{3}$

$a_3 = \frac{1}{3} * \frac{10}{3} = \frac{10}{9}$ $$$$
y asi sucesivamente.

La formula para $a_n$ podria escribir entonces como:

$a_n = 10 * (\frac{1}{3})^{n-1}$ $$$$

Anotacion:
Una secuencia geometrica es una secuencia de numeros donde cada termino despues del primero se obtiene multiplicando el termino anterior por una constante llamada razon comun r. La formula general para el n-esimo termino de una secuencia geometrica es:

$a_n = a * r^{n-1}$

r= razon comun.

a = primer termino.

```python
# @title $9)$ Consider the sequence {𝑎𝑛} with $a_1 = 2 $ and $ a_{n+1}=2a_n $  $$$$ for 𝑛 = 1, 2, 3, .. Plot the points (𝑛, 𝑎𝑛) for 𝑛 = 1, 2, 3, ... , 50.  Can you find thelimit of this sequence using the same method you used in Problem 8? Does this sequence have a limit?


import matplotlib.pyplot as plt

# Definir el número de términos a calcular
num_terms = 50

# Inicializar la secuencia
a_n = [2]  # a_1 = 2

# Calcular los términos siguientes
for n in range(1, num_terms):
    a_n.append(2 * a_n[-1])

# Definir los valores de n
n_values = list(range(1, num_terms + 1))

# Graficar la secuencia
#plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n, '.', color='blue', label='$a_n = 2 \cdot 2^{n-1}$')
plt.grid(True)
plt.legend()
plt.yscale('log')  # Escala logarítmica para visualizar mejor el crecimiento
plt.show()
```

$a_{n+1} = 2a_n$

primer termino a=2.

la razon comun r=2.

el termino n-esimo de la secuencia es:

$a_n = 2 * 2^{n-1} = 2^n$
