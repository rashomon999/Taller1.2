```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Crear una figura y una cuadrÃ­cula de subgrÃ¡ficas
fig, axs = plt.subplots(4, 3, figsize=(18, 16))

# GrÃ¡fico (a): (1/3)^n
n_values = list(range(20))  # Valores de n de 0 a 19
a_n = [1/3**n for n in n_values]  # CÃ¡lculo de la sucesiÃ³n
axs[0, 0].plot(n_values, a_n, '.', label='(a) SucesiÃ³n (1/3)^n')
axs[0, 0].set_title('(a) SucesiÃ³n (1/3)^n')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('a_n')
axs[0, 0].axhline(y=0, color='r', linestyle='-', label='LÃ­mite y=0')
axs[0, 0].legend()

# GrÃ¡fico (b): Ï^n
n_values = list(range(20))  # Valores de n de 0 a 19
a_n = [math.pi**n for n in n_values]  # CÃ¡lculo de la sucesiÃ³n
axs[0, 1].plot(n_values, a_n, '.', label='(b) SucesiÃ³n Ï^n')
axs[0, 1].set_yscale('log')  # Usar escala logarÃ­tmica para el eje y
axs[0, 1].set_title('(b) SucesiÃ³n Ï^n')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('a_n')
axs[0, 1].legend()

# GrÃ¡fico (c): (3n-1)/(2n+6) con lÃ­mite y=1.5
n_values = list(range(1, 60))  # Valores de n de 1 a 60
a_n = [(3*n-1)/(2*n+6) for n in n_values]  # CÃ¡lculo de la sucesiÃ³n
axs[0, 2].plot(n_values, a_n, '.', label='(c) SucesiÃ³n (3n-1)/(2n+6)')
axs[0, 2].axhline(y=1.5, color='r', linestyle='-', label='LÃ­mite y=1.5')
axs[0, 2].set_title('(c) SucesiÃ³n (3n-1)/(2n+6)')
axs[0, 2].set_xlabel('n')
axs[0, 2].set_ylabel('a_n')
axs[0, 2].legend()

# GrÃ¡fico (d): cos(Ï*n)
n_values = list(range(20))
a_n = [math.cos(math.pi*n) for n in n_values]
axs[1, 0].plot(n_values, a_n, '.', label='(d) SucesiÃ³n cos(Ï*n)')
axs[1, 0].set_title('(d) SucesiÃ³n cos(Ï*n)')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('a_n')
axs[1, 0].legend()

# GrÃ¡fico (e): (3n^4 - 6n^3 + 1) / (5n^3 + n^2 + 2)
n_values = list(range(1, 50))
a_n = [(3*n**4 - 6*n**3 + 1) / (5*n**3 + n**2 + 2) for n in n_values]  # CÃ¡lculo de la sucesiÃ³n
axs[1, 1].plot(n_values, a_n, '.', label='(e) (3n^4 - 6n^3 + 1) / (5n^3 + n^2 + 2)')
axs[1, 1].set_title('(e) SucesiÃ³n (3n^4 - 6n^3 + 1) / (5n^3 + n^2 + 2)')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('a_n')
axs[1, 1].legend()

# GrÃ¡fico (f): (2n^5 - 3n^2 + 23) / (7n^5 + 13n^4 - 12)
n_values = list(range(1, 20))
a_n = [(2*n**5 - 3*n**2 + 23)/(7*n**5 + 13*n**4 - 12) for n in n_values]
axs[1, 2].plot(n_values, a_n, '.', label='(f) (2n^5 - 3n^2 + 23) / (7n^5 + 13n^4 - 12)')
axs[1, 2].axhline(y=0.28571428571, color='r', linestyle='-', label='LÃ­mite y=0.28571428571')
axs[1, 2].set_title('(f) SucesiÃ³n (2n^5 - 3n^2 + 23) / (7n^5 + 13n^4 - 12)')
axs[1, 2].set_xlabel('n')
axs[1, 2].set_ylabel('a_n')
axs[1, 2].legend()

# GrÃ¡fico (g): (45 - 16n^2) / (13 + 5n + 6n^3)
n_values = list(range(1, 20))
a_n = [(45 - 16*n**2) / (13 + 5*n + 6*n**3) for n in n_values]
axs[2, 0].plot(n_values, a_n, '.', label='(g) (45 - 16n^2) / (13 + 5n + 6n^3)')
axs[2, 0].axhline(y=0, color='r', linestyle='-', label='LÃ­mite y=0')
axs[2, 0].set_title('(g) SucesiÃ³n (45 - 16n^2) / (13 + 5n + 6n^3)')
axs[2, 0].set_xlabel('n')
axs[2, 0].set_ylabel('a_n')
axs[2, 0].legend()

# GrÃ¡fico (h): (3n + 1) / â(4n^2 + 1)
n_values = np.linspace(1, 50, 300)  # Usamos linspace para obtener mÃ¡s puntos entre 1 y 50
a_n = (3*n_values + 1) / np.sqrt(4*n_values**2 + 1)
axs[2, 1].plot(n_values, a_n, '.', color='b', label='(h) SucesiÃ³n (3n+1) / â(4n^2 + 1)')
axs[2, 1].set_title('(h) SucesiÃ³n (3n+1) / â(4n^2 + 1)')
axs[2, 1].axhline(y=1.5, color='r', linestyle='-', label='LÃ­mite y=1.5')
axs[2, 1].set_xlabel('n')
axs[2, 1].set_ylabel('a_n')
axs[2, 1].legend()

# GrÃ¡fico (i): (-2)^(2n + 1)
n_values = list(range(1, 10))
a_n = [(-2)**(2*n + 1) for n in n_values]
axs[2, 2].plot(n_values, a_n, 'o', label='(i) SucesiÃ³n (-2)^(2n + 1)')
axs[2, 2].set_title('(i) SucesiÃ³n (-2)^(2n + 1)')
axs[2, 2].set_xlabel('n')
axs[2, 2].set_ylabel('a_n')
axs[2, 2].set_yscale('symlog')  # Escala logarÃ­tmica simÃ©trica en el eje y
axs[2, 2].legend()

# GrÃ¡fico (j): (10 - 16*n^3) / (1 + n^2)
n_values = list(range(1, 50))  # Generar valores de 1 a 49
a_n = [(10 - 16*n**3) / (1 + n**2) for n in n_values]
axs[3, 0].plot(n_values, a_n, '.', color='b', label='(j) SucesiÃ³n (10 - 16*n^3) / (1 + n^2)')
axs[3, 0].set_title('(j) SucesiÃ³n (10 - 16*n^3) / (1 + n^2)')
axs[3, 0].set_xlabel('n')
axs[3, 0].set_ylabel('a_n')
axs[3, 0].legend()

# GrÃ¡fico (k): â(3*n^2 + n - 6) / â(5*n^2 + 16)
n_values = list(range(2, 50))  # Generar valores de 1 a 49
a_n = [np.sqrt(3*n**2 + n - 6) / np.sqrt(5*n**2 + 16) for n in n_values]
axs[3, 1].plot(n_values, a_n, '.', color='b', label='(k) SucesiÃ³n â(3*n^2 + n - 6) / â(5*n^2 + 16)')
axs[3, 1].set_title('(k) SucesiÃ³n (3*n^2 + n - 6) / (5*n^2 + 16)')
axs[3, 1].axhline(y=0.77459666924, color='r', linestyle='-', label='LÃ­mite y=0.77459666924')
axs[3, 1].set_xlabel('n')
axs[3, 1].set_ylabel('a_n')
axs[3, 1].legend()

# GrÃ¡fico (l): (-1)**n) / (5**n)
n_values = list(range(0, 50))  # Generar valores de 0 a 50
a_n = [((-1)**n) / (5**n) for n in n_values]
axs[3, 2].plot(n_values, a_n, '.', color='b', label='(l) SucesiÃ³n (-1)**n) / (5**n)')
axs[3, 2].set_title('(l) SucesiÃ³n (-1)**n) / (5**n)')
axs[3, 2].axhline(y=0, color='r', linestyle='-', label='LÃ­mite y=0')
axs[3, 2].set_xlabel('n')
axs[3, 2].set_ylabel('a_n')
axs[3, 2].legend()

# Ajustar el espaciado entre grÃ¡ficos
plt.tight_layout()
plt.show()

```
